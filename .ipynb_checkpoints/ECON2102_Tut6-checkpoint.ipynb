```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root, minimize
plt.rcParams['lines.linewidth'] = '4'
plt.rcParams.update({'font.size': 16})
plt.rcParams['figure.figsize'] = (8,8)
```

```python
y1 = 100
y2 = 120
t1 = 20
t2 = 10
r  = 0.1
```

```python
c1 = np.linspace(0,200,1000)
c2 = (y1 - t1) * (1+r) + y2 -t2 - (1+r)*c1
```

```python
fig, ax = plt.subplots(1)

ax.plot(c1[:400], c2[:400], c = 'b')
ax.plot(c1[400:], c2[400:], c = 'r')
ax.set_xlim(0,200)
ax.set_ylim(0,200)
ax.axvline(x = y1-t1,ymin =0, ymax = (y2-t2)/200, c = 'k', linestyle = '--', linewidth = '2')
ax.axhline(y = y2-t2,xmin =0, xmax = (y1-t1)/200, c = 'k', linestyle = '--', linewidth = '2')
ax.set_xticks([80, 180])
ax.set_yticks([0,110, 198])
ax.text(40, 160,'      Saver, \nwhen c1 < 80', fontsize = 20, c = 'b')
ax.text(120, 70,'   Bowrrower, \nwhen c1 > 80', fontsize = 20, c = 'r')
ax.text(80, 120,'Slope = -(1+r) = -1.1')
ax.set_xlabel('c1')
ax.set_ylabel('c2')
```
