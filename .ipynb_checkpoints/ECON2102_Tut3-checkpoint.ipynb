```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root, minimize
plt.rcParams['lines.linewidth'] = '4'
plt.rcParams.update({'font.size': 16})
plt.rcParams['figure.figsize'] = (10,10)
```

```python
# Define the utility function
u = lambda c, l: c**(2/3) * l**(1/3)

## Find the optimal bundle (optimal consumption and optimal leisure)
# two conditions for tangent line
def FindTangent(x, u0):
    '''
    x: leisure and consumption
    u0: utility level needed to be attianed
    '''
    l = x[0]
    c = x[1]
    con1 = u0 - u(c, l)
    con2 = -1/2*u0**(3/2)*l**(-3/2) + 2
    return abs(con1) + abs(con2)


## Find the indifference curve
def Ind(c, l, opc, opl):
    return u(c,l) - u(opc, opl)
```

```python
#### part a

# Grid of x-axis (leisure)
l = np.linspace(0,1, 100)

# Budget constraint
ca = 1.2 - l

# optimal bundle for part a
opca = 0.8
opla = 0.4

## Find the indifference curve
la = np.linspace(0.15,1,100)
# Inda: indifference curve part a
Inda = root(Ind, x0 = np.ones(len(la)), args = (la, opca, opla)).x

# set range of axes
xlima = 1.2
ylima = 1.4

fig, ax = plt.subplots(1)
ax.plot(l,ca, label = 'Budget Constraint')
ax.plot(la,Inda, label = 'Indifference Curve')
ax.set_xlim(0,xlima)
ax.set_ylim(0,ylima)
ax.axvline(x=1, ymin =0, ymax = 0.2/ylima)
ax.axvline(x = opla, ymin = 0, ymax = opca/ylima, c = 'k', linestyle = '--', linewidth = '2')
ax.axhline(y = opca, xmin = 0, xmax = opla/xlima, c = 'k', linestyle = '--', linewidth = '2')
ax.scatter(0.4,0.8,s = 200,c = 'y')
ax.set_xlabel('Leisure l')
ax.set_ylabel('Consumption C')
ax.text(opla+0.05, opca,'({:1.2f}, {:1.2f})'.format(opla, opca))
ax.legend()
```

```python
## part b

cb = 2.2 - l

## current utility level
opcb = 4.4/3
oplb = 2.2/3

lb = np.linspace(0.27, 1, 100)
Indb = root(Ind, x0 = np.ones(len(lb)), args = (lb, opcb, oplb)).x

xlimb = 1.5
ylimb = 2.5


fig, ax = plt.subplots(1)

ax.plot(l,cb, label = 'Budget Constraint')
ax.plot(lb,Indb, label = 'Indifference Curve')
ax.set_xlim(0,1.5)
ax.set_ylim(0,2.5)
ax.axvline(x=1, ymin =0, ymax = 1.2/2.5)
ax.axvline(x = oplb, ymin = 0, ymax = opcb/ylimb, c = 'k', linestyle = '--', linewidth = '2')
ax.axhline(y = opcb, xmin = 0, xmax = oplb/xlimb, c = 'k', linestyle = '--', linewidth = '2')
ax.scatter(oplb,opcb,s = 200,c = 'y')
ax.set_xlabel('Leisure l')
ax.set_ylabel('Consumption C')
ax.text(oplb+0.05,opcb,'({:1.2f}, {:1.2f})'.format(oplb, opcb))
ax.legend()


```

```python
## part c


cc = 2.2 - 2*l

## current utility level
opcc = 4.4/3
oplc = 2.2/6


lc = np.linspace(0.15,1, 100)
Indc = root(Ind, x0 = np.ones(len(lc)), args = (lc, opcc, oplc)).x

xlimc = 1.2
ylimc = 2.4


fig, ax = plt.subplots(1)

ax.plot(l,cc, label = 'Budget Constraint')
ax.plot(lc,Indc, label = 'Indifference Curve')
ax.set_xlim(0,xlimc)
ax.set_ylim(0,ylimc)
ax.axvline(x=1, ymin =0, ymax = 0.2/2.5)
ax.axvline(x = oplc, ymin = 0, ymax = opcc/ylimc, c = 'k', linestyle = '--', linewidth = '2')
ax.axhline(y = opcc, xmin = 0, xmax = oplc/xlimc, c = 'k', linestyle = '--', linewidth = '2')
ax.scatter(oplc,opcc,s = 200,c = 'y')
ax.set_xlabel('Leisure l')
ax.set_ylabel('Consumption C')
ax.text(oplc+0.05,opcc,'({:1.2f}, {:1.2f})'.format(oplc, opcc))
ax.legend()
```

```python
## part c

plt.rcParams['lines.linewidth'] = '3'

xlim = max(xlima, xlimc)
ylim = max(ylima, ylimc)



fig, ax = plt.subplots(1)
ax.set_xlim(0,xlim)
ax.set_ylim(0,ylim)

ax.set_xlabel('Leisure l')
ax.set_ylabel('Consumption C')


ax.plot(l,ca, c = 'b')
ax.plot(la,Inda, c = 'r')
ax.axvline(x=1, ymin =0, ymax = 0.2/ylim, c = 'b')
ax.axvline(x = opla, ymin = 0, ymax = opca/ylim, c = 'k', linestyle = '--', linewidth = '2')
ax.axhline(y = opca, xmin = 0, xmax = opla/xlim, c = 'k', linestyle = '--', linewidth = '2')
ax.scatter(opla,opca,s = 100,c = 'y')
ax.text(opla+0.03,opca,'A')


ax.plot(l,cc, c = 'g')
ax.plot(lc,Indc, c = 'r')
ax.axvline(x = oplc, ymin = 0, ymax = opcc/ylim, c = 'k', linestyle = '--', linewidth = '2')
ax.axhline(y = opcc, xmin = 0, xmax = oplc/xlim, c = 'k', linestyle = '--', linewidth = '2')
ax.scatter(oplc,opcc,s = 100,c = 'y')
ax.text(oplc+0.03,opcc,'C')




x0 = minimize(FindTangent, x0 = [opla, opca], bounds = [(1e-10, 1-1e-10), (1e-10, 2.5)], args = u(opca, opla)).x

oplc0 = x0[0]
opcc0 = x0[1]
cc0 = opcc0 + 2* oplc0 - 2 * l


ax.scatter(oplc0,opcc0, s = 100, c = 'r')
ax.text(oplc0+0.03,opcc0,'C*')
ax.axvline(x = oplc0, ymin = 0, ymax = opcc0/ylim, c = 'k', linestyle = '--', linewidth = '2')
ax.axhline(y = opcc0, xmin = 0, xmax = oplc0/xlim, c = 'k', linestyle = '--', linewidth = '2')

ax.plot(l,cc0, c = 'g', linestyle = '-.')
ax.axvline(x = 0.029, ymin = opca/ylim, ymax = opcc0/ylim, c = 'm',linewidth = '8', label = 'Substitution Effect')
ax.axvline(x = 0.029, ymin = opcc0/ylim, ymax = opcc/ylim, c = 'y',linewidth = '8', label = 'Income Effect')
ax.axvline(x = 0.012, ymin = opcc/ylim, ymax = opca/ylim, c = 'c',linewidth = '8', label = 'Total Effect')
ax.axhline(y = 0.018, xmin = oplc0/xlim, xmax = oplc/xlim, c = 'y',linewidth = '8')
ax.axhline(y = 0.054, xmin = oplc0/xlim, xmax = opla/xlim, c = 'm',linewidth = '8')
ax.axhline(y = 0.018, xmin = oplc/xlim, xmax = opla/xlim, c = 'c',linewidth = '8')



ax.legend()
    
```

```python
# Part d
plt.rcParams.update({'font.size': 20})


w = np.array([0.5, 1, 1.5, 2])
l = (w + 0.2)/(3*w)
Ns = 1 - l


fig, ax = plt.subplots(1)

xlimd = 0.64
xmin = 0.52
ylimd = 2.2

ax.plot(Ns,w)
ax.set_xlabel('Labor supply Ns')
ax.set_ylabel('Wage w')
ax.scatter(Ns, w, c = 'r', s = 100)
ax.set_ylim(0, ylimd)
ax.set_xlim(xmin, xlimd)

for i in range(len(w)):
    ax.axvline(x = Ns[i], ymin = 0, ymax = w[i]/ylimd, c = 'k', linestyle = '--',linewidth = '2')
    ax.axhline(y = w[i], xmin = 0, xmax = (Ns[i]-xmin)/(xlimd - xmin), c = 'k', linestyle = '--',linewidth = '2')
```
