```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root, minimize
```

```python
u = lambda c, l: c**(2/3) * l**(1/3)
```

```python
## part a

plt.rcParams['lines.linewidth'] = '4'
plt.rcParams.update({'font.size': 16})
plt.rcParams['figure.figsize'] = (10,10)

l = np.linspace(0,1, 100)
c = 1.2 - l

css = 0.8
lss = 0.4

## current utility level
u0 = u(css,lss)
def CalC(c, l):
    return u(c,l) - u0

l0 = np.linspace(0.1,1, 100)
c0 = root(CalC, x0 = np.ones(len(l0)), args = l0).x

fig, ax = plt.subplots(1)

ax.plot(l,c, label = 'Budget Constraint')
ax.plot(l0,c0, label = 'Indifference Curve')
ax.set_xlim(0,1.2)
ax.set_ylim(0,1.5)
ax.axvline(x=1, ymin =0, ymax = 0.2/1.5)
ax.axvline(x = lss, ymin = 0, ymax = css/1.5, c = 'r', linestyle = '--')
ax.axhline(y = css, xmin = 0, xmax = lss/1.2, c = 'r', linestyle = '--')
ax.scatter(0.4,0.8,s = 200,c = 'y')
ax.set_xlabel('Leisure l')
ax.set_ylabel('Consumption C')
ax.text(lss+0.05,css,'({:1.2f}, {:1.2f})'.format(lss, css))
ax.legend()
```
